```python
corrected = corrected.drop([0])
```

```python
def addcorrect(row):
    global corrected
    if row[1] in ['Мужчина', 'Male', 'Mees']:
        row[1]='Male'
    if row[1] in ['Женщина', 'Female', 'Naine']:
        row[1]='Female'
    if row[0] == 'Русский':
        row[0] = 'Russian'
    if row[0] == 'Eesti':
        row[0] = 'Estonian'
    corrected = corrected.append({'Language': row[0], 
                                  'Sex': row[1],
                                  'Age': row[2],
                                  '30sec': row[3],
                                  '1min': row[4],
                                 '1min30sec': row[5],
                                 '2min': row[6]}, ignore_index=True)
```

```python
def makerow(df,row):
    newrow=[]
    for i in range(7):
        newrow.append(df.iat[row,i])
    return newrow
```

```python
for i in range(len(ru_data)):
    addcorrect(makerow(ru_data, i))
```

```python
for i in range(len(eng_data)):
    addcorrect(makerow(eng_data, i))
```

```python
def errors(data,interval,col):
    error=[]
    for i in range(len(data)):
        error.append(data.iat[i,col] - interval)
        
    return error
```

```python
df.insert(4,"30sec_error", errors(df,30,3))
df.insert(6,"1min_error", errors(df,60,5))
df.insert(8,"1min30sec_error", errors(df,90,7))
df.insert(10,"2min_error", errors(df,120,9))
```

```python
def increase():
    increase=[]
    for i in range(len(df)):
        if (abs(df.iat[i,4]) < abs(df.iat[i,6]) and abs(df.iat[i,6]) < abs(df.iat[i,8]) and abs(df.iat[i,8]) < abs(df.iat[i,10])):
            increase.append(True)
        else:
            increase.append(False)
    return increase
            
```

```python
import copy
def incignored():
    increase=[]
    for i in range(len(df)):
        errors=[]
        for j in range(4,12,2):
            errors.append(df.iat[i,j])
        increased = False
        for n in range(4):
            testing = errors.copy()
            testing.pop(n)
            if abs(testing[0]) < abs(testing[1]) and abs(testing[1]) < abs(testing[2]):
                increase.append(True)
                increased = True
                break
        if increased == False:  
            increase.append(False)
                
    return(increase)
```

```python
def error_sum():
    summ = []
    for i in range(len(df)):
        error_sum = 0
        for j in range(4,12,2):
            error_sum += abs(df.iat[i,j])
        summ.append(error_sum)
    return summ
```

```python
def error_mean():
    mean = []
    for i in range(len(df)):
        error_sum = 0
        for j in range(4,12,2):
            error_sum += abs(df.iat[i,j])
        mean.append(error_sum/4)
    return mean
```
